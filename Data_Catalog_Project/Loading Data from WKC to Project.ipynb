## Loading data from Watson Knowledge Catalog to Watson Studio Project

## Overview and Prerequisites

This notebook is a continuation of Data Acquisiton section. We will be importing the virtualized dataset, `Airline_Virtualized_Data_User_Name` into a Watson Studio Project.

## Prerequisites

This notebook assumes you have Virtualized the data using instructions provided in `Data Virtualization (Postgres + DB2)` notebook.

## Loading data to the project

### 1. Starting from the CP4D home screen, navigate to the Sandwich icon at the top left corner, then go to 'Catalogs' > 'All catalogs'. This will take you to a view of your available data catalogs. Click on the catalog assigned to the user - `Model-Catalog-User-name` Catalog. 



### 2. Click on the `Airline_Virtualized_Data_User-name` dataset. Then click "Add to Project +" on the top right corner.This will take you to the "Add to project" page




### 3. On the "Add to project" page, select the `Airline-MLOps-Dev-User-Name` project where you want to add the dataset  and then click on "Add" at the bottom right corner. 





### 4. Navigate to the Sandwich icon at the top left corner, then go to 'Projects' > 'All projects'. This will take you to a view of your available Watson Studio projects. 

### 5. Click on the Airline-MLOps-Dev-`User-Name` project where you added the dataset. You will be able to see the virtualized dataset under the "Data Assets" section.
